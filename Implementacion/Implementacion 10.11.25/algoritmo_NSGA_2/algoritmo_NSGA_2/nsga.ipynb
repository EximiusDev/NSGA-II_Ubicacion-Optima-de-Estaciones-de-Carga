```python
# EJEMPLO DE COMO CREAR UN PROBLEMA PARA APLICAR A NSGA-2

from pymoo.core.problem import ElementwiseProblem
import numpy as np

# Definís tu problema personalizado
class MyEVProblem(ElementwiseProblem):

    def __init__(self):
        super().__init__(
            n_var=3,          # Número de variables de decisión
            n_obj=2,          # Número de funciones objetivo (f1, f2)
            n_constr=0,       # Número de restricciones (podés poner >0 si las necesitás)
            xl=np.array([0, 0, 0]),   # Límite inferior de cada variable
            xu=np.array([1, 5, 10])   # Límite superior de cada variable
        )

    def _evaluate(self, x, out, *args, **kwargs):
        # Definís tus funciones objetivo f1 y f2
        f1 = x[0]**2 + x[1]**2 + x[2]**2         # por ejemplo: coste
        f2 = (x[0]-1)**2 + (x[1]-2)**2 + (x[2]-3)**2  # por ejemplo: distancia promedio

        # Podés agregar restricciones si querés
        # g1 = x[0] + x[1] - 2
        # out["G"] = [g1]

        out["F"] = [f1, f2]
```

```python
# EJEMPLO DE COMO IMPLEMENTAR NSGA-2 AL PROBLEMA CREADO MANUALMENTE

from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter

# Crear una instancia de tu problema
problem = MyEVProblem()

# Definir el algoritmo
algorithm = NSGA2(pop_size=100)

# Ejecutar la optimización
res = minimize(problem,
               algorithm,
               ('n_gen', 200),
               seed=1,
               verbose=True)

# res.F → matriz con los valores de las funciones objetivo [f1, f2] para cada solución.
# res.X → matriz con los valores de las variables de decisión correspondientes.

# Graficar la frontera encontrada
plot = Scatter()
plot.add(res.F, color="red")
plot.show()
# El gráfico muestra la frontera de Pareto (los compromisos entre tus dos métricas).
```
